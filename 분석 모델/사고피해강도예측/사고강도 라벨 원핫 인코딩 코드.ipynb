```python
계절=[]#월을 계절로
for i in data['월']:
        if i>=3 and i<=5:
            계절.append("봄")
        elif i>=6 and i<=8:
            계절.append("여름")
        elif i>=9 and i<=11:
            계절.append("가을")
        else:
             계절.append("겨울")
```

```python
data['도로형태'].str.split('-')#도로형태를 단일로 교차로 기타로 구분
```

```python
도로구분=[]#도로형태
도로구분1=[]
for i in range(len(data['도로형태'])):
    도로구분.append(data['도로형태'].str.split('-')[i][0])
    도로구분1.append(data['도로형태'].str.split('-')[i][1])
```

```python
data['계절']=계절
data['도로구분']=도로구분
data['도로구분1']=도로구분1
#각 새로운 컬럼 생성
```

```python
포장구분=[]#도로형태
노면구분=[]#노면구분
for i in range(len(data['노면상태'])):
    포장구분.append(data['노면상태'].str.split('-')[i][0])
    노면구분.append(data['노면상태'].str.split('-')[i][1])
```

```python
data['포장구분']=포장구분
data['노면구분']=노면구분
```

```python
eud1=pd.get_dummies(data["계절"])#범주형인 계절 데이터 원핫인코딩
```

```python
eud2=pd.concat([edu_data,eud1],axis=1)#원데이터는 원핫데이터 범주와 합치기
```

```python
eud2#합친데이터 확인
```

```python
eud3=pd.get_dummies(data["도로구분"])# 도로구분을 원핫 인코딩 합니다
encoding_data=pd.concat([eud2,eud3],axis=1)#원핫과 본데이터를 합치기
encoding_data
```

```python
encoding_data=encoding_data.drop(['계절','도로구분','피해운전자 상해정도'],axis=1)# 범주형 데이터가 있는 컬럼제거
```

```python
encoding_data["피해운전자 상해정도"]=data['피해운전자 상해정도']#데이터 상해정도 입력
```

```python
encoding_data
```

```python
encoding_data['가해운전자 차종'].value_counts()#가해 운전자 차종 확인
```

```python
encoding_data.loc[encoding_data['가해운전자 차종']=='자전거','가해운전자 차종']='이륜'
encoding_data.loc[encoding_data['가해운전자 차종']=='개인형이동수단(PM)','가해운전자 차종']='이륜'
encoding_data.loc[encoding_data['가해운전자 차종']=='화물','가해운전자 차종']='특수차'
encoding_data.loc[encoding_data['가해운전자 차종']=='원동기','가해운전자 차종']='특수차'
encoding_data.loc[encoding_data['가해운전자 차종']=='건설기계','가해운전자 차종']='특수차'
encoding_data.loc[encoding_data['가해운전자 차종']=='기타불명','가해운전자 차종']='기타차량'
encoding_data.loc[encoding_data['가해운전자 차종']=='특수','가해운전자 차종']='기타차량'
## 자전거 개인형 화물수단은 이륜이기 때문에 이륜으로 지정
## 화물 원동기 건설기계는 특수로 묶어 하나의 차량으로 지정
## 기타불명 특수는 특수차의 범위가 크기 때문에 기타차량으로 지정하였습니다
```
