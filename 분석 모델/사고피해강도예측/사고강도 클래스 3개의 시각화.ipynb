```python
import pandas as pd #판다스 호출
import numpy as np #넘파이 호출
import matplotlib.pyplot as plt#파이 플롯 호출
import seaborn as sns #seaborn 호출
%matplotlib inline
plt.rcParams['font.family'] = 'NanumGothic'#한글깨짐 방지
```

```python
data=pd.read_csv("데이터1.csv",encoding='cp949')#데이터 호출
```

```python
data=data.drop('Unnamed: 0',axis=1)#불필요한 데이터제거
```

```python
data.head()#데이터 확안
```

```python
data['도로구분'].unique()#특정 데이터의 데이터라벨 확인 
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(x='피해운전자 상해정도',hue='요일',data=data,orient="v")#요일에 따른 상해정도 확인
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 3, height, ha = 'center', size = 9)

plt.show()
```
