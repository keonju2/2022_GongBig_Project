```python
import pandas as pd #판다스 호출
import numpy as np # 넘파이 호출
import matplotlib.pyplot as plt #파이플롯 호출
import seaborn as sns #seaborn 호출
%matplotlib inline
plt.rcParams['font.family'] = 'NanumGothic'#한글 깨지기 방지
```

```python
data=pd.read_csv("데이터.csv",encoding='cp949')# 데이터 호출
```

```python
data=data.drop('Unnamed: 0',axis=1) # 불필요한 컬럼 제거
```

```python
data.head() # 데이터 앞 확인
```

```python
data.loc[(data['연령대']==10.0) & (data['피해운전자 상해정도']=='경상')].count()
```

```python
data.loc[data['가해운전자 차종']=='자전거','가해운전자 차종']='이륜'
data.loc[data['가해운전자 차종']=='개인형이동수단(PM)','가해운전자 차종']='이륜'
data.loc[data['가해운전자 차종']=='화물','가해운전자 차종']='특수차'
data.loc[data['가해운전자 차종']=='원동기','가해운전자 차종']='특수차'
data.loc[data['가해운전자 차종']=='건설기계','가해운전자 차종']='특수차'
data.loc[data['가해운전자 차종']=='기타불명','가해운전자 차종']='기타차량'
data.loc[data['가해운전자 차종']=='특수','가해운전자 차종']='기타차량'
#차량 범주 축소
```

```python
data['가해운전자 차종'].value_counts()
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(x='피해운전자 상해정도',hue='요일',data=data,orient="v")#요일별 피해운전자 상해정도
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 3, height, ha = 'center', size = 9)

plt.show()
```
