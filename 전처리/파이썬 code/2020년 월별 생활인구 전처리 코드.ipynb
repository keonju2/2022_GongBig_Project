```python
data_1=pd.DataFrame()
data_2=pd.DataFrame()
data_3=pd.DataFrame()
data_4=pd.DataFrame()
#하나의 데이터프레임으로 묶기에 각 파일이 커서 데이터 프레임을 4개로 만들어 4개의 데이터 프레임을 합친다
```

```python
file1=files[:15]
file2=files[15:24]
file3=files[24:27]
file4=files[27:]
#전체 파일을 슬라이싱하여 4개로 나누어 합치기 위해 나눈다 
```

```python
for i in file1:
    path=i
    add=pd.read_csv(path,encoding='cp949')
    data_1=pd.concat([data_1,add])
#data_1과 1~14까지의 파일을 합친다
```

```python
for i in file2:
    path=i
    add=pd.read_csv(path,encoding='cp949')
    data_2=pd.concat([data_2,add])
#data_2와 15~23의 파일을 합친다
```

```python
for i in file3:
    path=i
    add=pd.read_csv(path,encoding='cp949')
    data_3=pd.concat([data_3,add])
#data_3와 24~26의 파일을 합친다
```

```python
for i in file4:
    path=i
    add=pd.read_csv(path,encoding='cp949')
    data_4=pd.concat([data_4,add])
#data_4와 나머지의 파일을 합친다
```

```python
data_1=data_1.replace("*",0)
#데이터1의 결측값을 0으로 대체
```

```python
data_2=data_2.replace("*",0)
#데이터2의 결측값을 0으로 대체
```

```python
data_3=data_3.replace("*",0)
data_4=data_4.replace("*",0)
#데이터3와4의 결측값을 0으로 대체
```

```python
data_com1=pd.concat([data_1,data_2])
#data1과data2를 합친다
```

```python
data_com2=pd.concat([data_3,data_4])
#data3과data4를 합친다
```

```python
data=pd.concat([data_com1,data_com2])
#최종적으로 두데이터를 합쳐 총 하나의 월데이터를 만든다
```

```python
data.head()#데이터 확인
```

```python
data1=data.iloc[:,1:]#?"기준일ID" 제거
```

```python
data1.shape
```

```python
data1.head()#데이터 확인
```

```python
주간야간=[]#6~18시를 주간 19~5시를 야간 시간대를 주간야간으로 구분
for i in data1['시간대구분']:
    if i>=6 and i<=18:
        주간야간.append("주간")
    else:
        주간야간.append("야간")
        
```

```python
data1["행정동코드"]=data1["행정동코드"].astype('str')+'00'
#다른 행정동코드와 비교했을 때 00이 없었으므로 00추가
```

```python
data1['월']=11#월이 11일이기 때문에 11로 설정 해당월에 해당하는 월 기입
```

```python
data2=data1.astype('float64')# 그룹바이 사용위해 실수형으로 설정
```

```python
data2["주간야간"]=주간야간# 시간대를 주간야간으로 설정하기 위해
```

```python
data4=data2.groupby(['집계구코드','주간야간','월']).mean()#집계구별 주간야간의 월별 인구수 평균
```
