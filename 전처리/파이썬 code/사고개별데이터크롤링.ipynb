```python
'''
16년 무단횡단사고 다발지역
13~15년 무단횡단사고 다발지역
12~14년 무단횡단사고 다발지역
11~13년 무단횡단사고 다발지역
10~12년 무단횡단사고 다발지역
09~11년 무단횡단사고 다발지역
08~10년 무단횡단사고 다발지역
07~09년 무단횡단사고 다발지역
'''
```

```python
from selenium import webdriver
import time
driver = webdriver.Chrome() #크롬드라이버실행
url = 'http://taas.koroad.or.kr/gis/mcm/mcl/initMap.do?menuId=GIS_GMP_AGS_TMM#' #URL
driver.get(url)
driver.implicitly_wait(10) 

driver.find_element('id','menuAccidentHazardSearch').click() #무단횡단 사고다발 클릭, 지도기다리기
driver.implicitly_wait(10) 
time.sleep(3)

driver.find_element('xpath',"//*[@id='tasPafItem']/option[text()='무단횡단사고 다발지역']").click() # 무단횡단 클릭
time.sleep(3)
year = '07~09년 무단횡단사고 다발지역'

driver.find_element('xpath',"//*[@id='tasPafGroup']/option[text()='{연도}']".format(연도=year)).click() #연도선택
time.sleep(3)

driver.find_element('xpath','//*[@id="roadAccidentCauseSectionFind"]/div[1]/fieldset/p/a').click() # 전체 기다리기
time.sleep(20)
```

```python
total = driver.find_element('xpath','//*[@id="roadAccidentCauseSectionFindSection"]/div[1]/div[1]/span').text
print(total) # 전체 몇개인지

page=1
name_list = []
page_list = []
while page<int(total):
    for i in range(1,11):
        try:
            driver.find_element('xpath','//*[@id="csfAcdntInfoZoneList"]/li[{i}]'.format(i=i)).click() # 사고세부내용 클릭

        except:
            driver.find_element('xpath','//*[@id="csfAcdntInfoZoneList"]/li[{i}]'.format(i=i)).click()

        time.sleep(10)
        driver.find_element('xpath','//*[@id="chicken_contentDiv"]/div[2]/div/div/div/p/a').click() #개별 데이터 정보 획득
        name = driver.find_element('xpath','//*[@id="chicken_contentDiv"]/div[2]/div/table/tbody/tr[1]/td[2]').text
        name_list.append(name)
        page_list.append(page)
        time.sleep(10)
        driver.switch_to.window(driver.window_handles[-1])
        driver.find_element('xpath','/html/body/div/input').click() # 개별데이터 새 창 클릭
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        page = page+1
    driver.find_element('xpath','//*[@id="roadAccidentCauseSectionFindSectionPagination"]/li[6]/a').click() # 엑셀 데이터 다운로드
```

```python
import pandas as pd
total = pd.DataFrame({'num':page_list,'name':name_list}) # 파일명이 같기 때문에 사고지역 이름 획득
```
