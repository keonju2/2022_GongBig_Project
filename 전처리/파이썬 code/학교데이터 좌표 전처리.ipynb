```python
import json
import requests

api_key = '' 
```

```python
def addr_to_lat_lon(addr): # 카카오 api를 통해 좌표 반환
  url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(addr)
  headers = {
      "Authorization": "KakaoAK "
  }
  places = requests.get(url, headers = headers).json()
  x=places['documents'][0]['x']
  y=places['documents'][0]['y']
  point=[y,x]
  return point
```

```python
latlon = [] # 학교 검색후 좌표 저장
for i in univ['학교명']:
  try:
    latlon.append(addr_to_lat_lon(i))
  except:
    latlon.append(None)
```

```python
univ['위경도'] = latlon
```
