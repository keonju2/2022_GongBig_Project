```python
import pandas as pd# 판다스호출

import numpy as np#넘파이 호출
import os
from glob import glob#글롭 호출
```

```python
data_1=pd.DataFrame()#데이터프레임 생성
```

```python
files=glob('C:/Users/user/Desktop/2020년 주간야간/*.csv')
files#20년 월별 생활인구수 파일 호출
```

```python
cd C:/Users/user/Desktop/2020년 주간야간
#디렉토리 변경
```

```python
for i in files:
    path=i
    add=pd.read_csv(path,encoding='cp949')
    data_1=pd.concat([data_1,add])
#데이터 합치기
```

```python
data_1
#데이터 호출
```

```python
data_1=data_1.drop(['Unnamed: 0','index'],axis=1)
#데이터 컬럼 제거
```

```python
data_1.reset_index(inplace=True)
#인덱스 재생성
```

```python
data_1#데이터 확인 
```

```python
data_1.drop(['index'],axis=1,inplace=True)
#인덱스 컬럼 제거
```

```python
data_1#데이터 최종확인
```

```python
data2=data_1.groupby(['집계구코드','주간야간']).mean()#각 집계구별 년도별 주간야간 평균 구하기
```

```python
data2.reset_index(inplace=True)#인덱스 재생성
```

```python
data2#데이터 확인
```

```python
data2['집계구코드'].value_counts()#집계구 누락유무 확인
```

```python
data3=data2.drop('월',axis=1)#불필요한 컬럼 제거
```

```python
data3.to_csv('20년 평균 집계구 인원.csv',encoding='cp949')#데이터 저장
```
